```python
import MDAnalysis as mda
from prolif import fingerprint, molecule
```

### SLCO2A1 + P2E

```python
gro_file = f"/home/bioc1870/SLCO2A1_simulations/RatSLCO2A1_P2E/production/rep_1/md.gro"  # To get the correct residue numbering
tpr_file = f"/home/bioc1870/SLCO2A1_simulations/RatSLCO2A1_P2E/production/rep_1/md_system.tpr"
xtc_file = f"/home/bioc1870/SLCO2A1_simulations/RatSLCO2A1_P2E/analysis/md_all_center.xtc"

u = mda.Universe(tpr_file, xtc_file)
u_ref = mda.Universe(gro_file)
u.residues.resids = u_ref.residues.resids[:len(u.residues.resids)]  # Fix the residue numbering in u
ligand_sel = u.select_atoms("resname P2E")

fp = fingerprint.Fingerprint.from_pickle(f'results/P2E/P2E_ifp_all.pkl')
ligand_mol = molecule.Molecule.from_mda(ligand_sel,)
fp.plot_lignetwork(ligand_mol, threshold=0.5)
```
